### Plotting histograms
#### The reason I am doing this is to explore any cumulative frequency distribution. I choose the categories ''Discount_offered", "Weight_in_gms" and "Cost_of_the_Product" because these are numerical. This is also still part of business & data understanding. Before I move on, I want to understand the data better. Because I know beforehand that I want to use multiple graphs, we make use of the libraries we imported earlier. The subplot function is a method of the matplotlib library in Python that allows you to create multiple plots within a single figure. It divides the figure into a grid of subplots, where each subplot is an individual plot. This allows you to compare multiple plots side by side and create more complex visualizations.

```python
# We want to create multiple graphs so we use this function.
fig, axes = plt.subplots(3,1,figsize=(10,16))

# cumulative frequency distribution for 'Cost_of_the_Product'
sns.histplot(df.Cost_of_the_Product, kde=True, ax=axes[0])
ax=axes[0].set_title('Frequency Distribution of Product Cost',fontsize=13,color="red")
ax=axes[0].set_xlabel('Product Cost (in dollar)',fontsize=13,color="red")
ax=axes[0].set_ylabel('Count',fontsize=13,color="red")

# cumulative frequency distribution for 'Weight_in_gms'
sns.histplot(df.Weight_in_gms, kde=True, ax=axes[1])
ax=axes[1].set_title('Frequency Distribution of Weight',fontsize=13,color="green")
ax=axes[1].set_xlabel('Weight (in gms)',fontsize=13,color="green")
ax=axes[1].set_ylabel('Count',fontsize=13,color="green")

# cumulative frequency distribution for'Discount_offered'
sns.histplot(df.Discount_offered, kde=True, ax=axes[2])
ax=axes[2].set_title('Frequency Distribution of Discount Offered',fontsize=13,color="blue")
ax=axes[2].set_xlabel('Discount Offered (in %)',fontsize=13,color="blue")
ax=axes[2].set_ylabel('Count',fontsize=13,color="blue")
```

#### Based on the histograms above, we can make a couple of conclusions. The price of the majority of the products typically falls within a range of 240 to 275 dollars. Most of the products weigh between 1,000 and 2,000 gms or 4,000 and 6,000 gms. It is common for discounts of 1% to 10% to be offered, with some outliers falling in the upper quartile.

### Modelling
#### In the modelling fase, you want to specify what steps need to be taken in order to solve a particular problem.  

#### In order to confirm if the company really seems to have a problem with not having loyal customers, these are the steps I will take. Now that I understand the data better, It is important to really specify what I want to make a visualization of.  I will compare the amount of customers and amount of prior purchases they made. This will give us a view if the company has loyal customers or not. The next step would be to look into the data to find possible causes and dig deeper. 

```python
sns.countplot(x = "Prior_purchases", data = df)

plt.title("Number of Prior Purchases Made by Customers")

plt.show()
```

#### Based on the graphs, the company does seem to have a problem with retaining loyal customers. Most customers have made 3 prior purchases. After this, it keeps getting less and less. The company wants customers to keep returning to them when buying electronics and this doesnt seem to be the case. Lets find out what a possible reason for this could be. When I looked at their data frame in the data understanding section, a column that is very important these days is delivery time. I want to find out if they deliver their products on time or not. This could be a possible reason for customers not returning because nothing is worse that ordering something and not getting it on time.

```python
sns.countplot(x = "Reached.on.Time_Y.N", data = df)

plt.title("Number of orders based on arrival time")

plt.show()
```

#### In the data frame, 0 stands for Yes and 1 stands for no. What does seem to be the main reason for customers not returning is that more than half of the 10,999 orders out of the total amount were not delivered on time. So simply said, the solution would be to deliver them on time but this answer is too easy. So we will dig deeper. 

```python
sns.countplot(x = "Mode_of_Shipment", data = df)

plt.title("Number of orders by shipment mode")

plt.show()
```

#### Most of the orders were shipped by ship. This explains why it takes longer than usual to get the products delivered to the customers. In the graph below it also seems that most customers call customer care 4 times. This is never really a good sign because customers wont call the company unless something is wrong. With all of the CRISP-DM steps we took, we came to the conclusion that in order to have customers returning, the company needs to do something about their supply chain so they can deliver products on time. Because over half of the products they deliver, it is not in time and this is just too much in order to keep a good running business.
